```python
import pandas as pd
import numpy as np
```

```python

hdi = pd.read_csv('Datos/Finales/HDR21-22_Composite_indices_complete_time_series.csv')
```

```python
hdi.head()
```

```python
hdi.columns.to_list()
```

```python
# Crear una lista de columnas que contienen "hdi" en su nombre
hdi_cols = [col for col in hdi.columns if col.lower().startswith("hdi")]

# Crear un nuevo dataframe con las columnas seleccionadas
df_hdi = hdi.filter(hdi_cols, axis=1)
```

```python
df_hdi.head()
```

```python
df_hdi.columns.to_list()
```

```python
col_fijas = hdi.iloc[:, :5]
```

```python
col_fijas.head()
```

```python
hdi_concat=pd.concat([col_fijas,df_hdi],axis=1)
hdi_concat.head()
```

```python
#Eliminamos la columna duplicada
hdi_concat = hdi_concat.loc[:, ~hdi_concat.columns.duplicated()]
hdi_concat.head()
```

```python
hdi_total = hdi_concat.iloc[:, 5:37]
hdi_total.head()
```

```python
new_cols = {}
for col in hdi_concat.columns:
    if col.startswith('hdi_'):
        year = col.split('_')[-1]
        new_cols[col] = year

hdi_total = hdi_total.rename(columns=new_cols)
```

```python
hdi_f = hdi_concat.iloc[:, 37:69]
hdi_f.head()
```

```python
hdi_m = hdi_concat.iloc[:, 69:]
hdi_m.head()
```

```python
hdi_concat=pd.concat([col_fijas,hdi_total],axis=1)
hdi_concat.head()
```

```python
hdi_melt=hdi_concat.melt(id_vars=['iso3','country','hdicode','region','hdi_rank_2021'], var_name='Year')
hdi_melt.head()
```

```python
col_fijas.head()
```

```python
hdi_concat2=pd.concat([col_fijas,hdi_f],axis=1)

hdi_melt2=hdi_concat2.melt(id_vars=['iso3','country','hdicode','region','hdi_rank_2021'], var_name='Year_F')
hdi_melt2.head()
```

```python
hdi_f = hdi_melt2[['Year_F', 'value']].rename(mapper={'value': 'Value_F'}, axis=1)
```

```python
hdi_f
```

```python
hdi_concat3=pd.concat([col_fijas,hdi_m],axis=1)

hdi_melt3=hdi_concat3.melt(id_vars=['iso3','country','hdicode','region','hdi_rank_2021'], var_name='Year_M')
hdi_melt3.head()
```

```python
hdi_m = hdi_melt3[['Year_M', 'value']].rename(mapper={'value': 'Value_M'}, axis=1)
```

```python
hdi_m.head(1)
```

```python
hdi_f.head(2)
```

```python
hdi_melt.reset_index(inplace=True, drop=True)
hdi_f.reset_index(inplace=True, drop=True)
hdi_m.reset_index(inplace=True, drop=True)
```

```python
hdi_melt["Female"] = hdi_f["Value_F"]
hdi_melt["Male"] = hdi_m["Value_M"]
hdi_melt.head()
```
