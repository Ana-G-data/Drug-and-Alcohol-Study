```python
def rename_columns(df):
    for col in df.columns:
        # Busca el año en el nombre de la columna usando una expresión regular
        match = re.search(r'\d{4}', col)
        if match:
            # Extrae el año de la cadena que coincide con la expresión regular
            year = match.group()
            # Renombra la columna usando solo el año
            df.rename(columns={col: year}, inplace=True)
```

```python
rename_columns(alcohol_df)
```
