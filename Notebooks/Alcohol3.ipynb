```python
alcohol_filtrado = alcohol_df.loc[alcohol_df['Indicator Name'].str.contains('alcohol')]

alcohol_filtrado
```

```python
alcohol_filtrado = alcohol_filtrado.dropna(axis=1, how='all')
alcohol_filtrado.info()
```

```python
alcohol_filtrado=alcohol_filtrado.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year')
alcohol_filtrado.head()
```

```python
alcohol_filtrado = alcohol_filtrado.rename(columns={'Indicator Name': 'Total alcohol consumption per capita (liters of pure alcohol)'})
alcohol_filtrado.head()
```

```python
def gender_from_total_alcohol(text):
    if re.search(r'\bmale\b', text, re.IGNORECASE):
        return 'Male'
    elif re.search(r'\bfemale\b', text, re.IGNORECASE):
        return 'Female'
    else:
        return 'Both'
```

```python
alcohol_filtrado['Total alcohol consumption per capita (liters of pure alcohol)'] = alcohol_filtrado['Total alcohol consumption per capita (liters of pure alcohol)'].apply(gender_from_total_alcohol)
alcohol_filtrado.head(6)
```

```python
# crear una tabla dinámica con el país y el año como índices y el género como columnas
pivot_data = pd.pivot_table(alcohol_filtrado, index=['Country Name', 'Year'], columns='Total alcohol consumption per capita (liters of pure alcohol)', values='value')

# sumar las columnas de Male y Female y asignar el resultado a la columna Both
pivot_data['Both'] = pivot_data['Male'] + pivot_data['Female']

# resetear los índices para convertirlos en columnas
reset_data = pivot_data.reset_index()

# convertir las columnas de género en una sola columna
melt_data = reset_data.melt(id_vars=['Country Name', 'Year'], value_vars=['Female', 'Male', 'Both'], var_name='Gender', value_name='Value')

melt_data.head()


```

```python
melt_data.sort_values(by=(['Year','Country Name']))
```

```python
df_new = melt_data.set_index(['Country Name', 'Year', 'Gender'])['Value'].unstack('Gender').reset_index()
df_new = df_new.rename(columns={'Female': 'Value_Female', 'Male': 'Value_Male', 'Both': 'Value_Both'})
df_new=df_new.round(2)
df_new
```

```python
def get_iso3(country_name):
    try:
        country = pycountry.countries.get(name=country_name)
        return country.alpha_3
    except:
        return None
```

```python
df_new['ISO3'] = df_new['Country Name'].apply(get_iso3)
```
