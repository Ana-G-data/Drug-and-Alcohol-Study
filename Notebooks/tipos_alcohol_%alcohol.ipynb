```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
df = pd.read_csv("Datos/Finales/consumo_tipo_alcohol.csv",index_col=0)
df.head()
```

```python
df_new = df.set_index(['Continent','ISO3','Country', 'Year', 'Type'])['Total'].unstack('Type').reset_index()

df_new.head()
```

```python
df_new['Low Alcohol%']=df_new['Beer']+df_new['Wine']
df_new['High Alcohol%']=df_new['Spirits']+df_new['Other alcoholic beverages']
df_new.head()
```

```python
df_new.to_csv('Datos/Finales/consumo_%_alcohol.csv')
```
