```python
import pandas as pd
import re
```

```python
population_df=pd.read_csv('Datos/drogas/World population.csv')
population_df
```

```python
population_df.duplicated().sum()
```

```python
population_df.columns
```

```python
population_df=population_df[['ISO3_code', 'Location', 'Time', 'PopMale', 'PopFemale', 'PopTotal']]
population_df.head(2)
```

```python
population_df = population_df.rename(columns={
    'ISO3_code': 'ISO3',
    'Location': 'Country',
    'Time': 'Year',
})
population_df.head(2)
```

```python
population_df_final = population_df.loc[(population_df['Year'] >= 1998) & (population_df['Year'] <= 2021)].copy()
population_df_final
```

```python
population_df_final['Year'].unique
```

```python
df_grouped = population_df_final.groupby(['ISO3', 'Year', 'Country']).agg({'PopMale':'sum', 'PopFemale':'sum', 'PopTotal':'sum'}).reset_index()

df_grouped.head()
```

```python
df_grouped[df_grouped['Country'] == 'United States of America']
```

```python

df_grouped[['PopMale', 'PopFemale', 'PopTotal']] = df_grouped[['PopMale', 'PopFemale', 'PopTotal']].multiply(1000, axis=1)
df_grouped.head()
```

```python
df_grouped=df_grouped.to_csv('Datos/Finales/world_population.csv')
```
