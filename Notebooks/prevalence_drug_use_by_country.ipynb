```python
def get_iso3(country_name):
    try:
        country = pycountry.countries.get(name=country_name)
        return country.alpha_3
    except:
        return None
```

```python
df_drogas['ISO3'] = df_drogas['Country/Territory'].apply(get_iso3)
```

```python
df_drogas.isnull().sum()
```

```python
null_iso3 = df_drogas.loc[pd.isnull(df_drogas['ISO3']), 'Country/Territory'].unique()
print(null_iso3)
```

```python
iso3_dict = {}

for pais in null_iso3:
    try:
        codigo = pycountry.countries.search_fuzzy(pais)[0].alpha_3
        iso3_dict[pais] = codigo
    except LookupError:
        print(f"No se encontr el cdigo ISO3 para {pais}")

print(iso3_dict)
```

```python
iso3_manual = {
    'Bolivia (Plurinational State of)': 'BOL',
    'Venezuela (Bolivarian Republic of)': 'VEN',
    'China, Taiwan Province of China': 'TWN',
    'Iran (Islamic Republic of)': 'IRN',
    'Kosovo under UNSCR 1244': 'XKX',
    'Trkiye': 'TUR',
    'United Kingdom (England and Wales)': 'GBR',
    'United Kingdom (Northern Ireland)': 'GBR',
    'United Kingdom (Scotland)': 'GBR',
    'Democratic Republic of the Congo': 'COD'}
```

```python
iso3 = iso3_dict.copy()
iso3.update(iso3_manual)
```

```python
def convert_columns_to_float(df, cols_to_convert, num_decimals):
    for col in cols_to_convert:
        df[col] = df[col].apply(lambda x: round(pd.to_numeric(x, errors='coerce') * 10**num_decimals, 0) / 10**num_decimals if not np.isnan(x) else x)
```

```python
cols_to_convert = ['Total', 'Female', 'Male', 'Year', 'Low', 'High']
num_decimals = 2

convert_columns_to_float(df_drogas, cols_to_convert, num_decimals)
```
