```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
```

```python
ansiedad_porcentaje=pd.read_csv('Datos/ansiedad_pais_año.csv',index_col=0)
mentaldisorders_numero=pd.read_csv('Datos/salud mental/mentaldisorder_pais_año.csv',index_col=0)
depresion_numero=pd.read_csv('Datos/depressiondisorder_pais_1990-2019.csv',index_col=0)
```

```python
# merge por columna Country
enfermedades_totales = pd.merge(ansiedad_porcentaje, mentaldisorders_numero, on=['Entity','Year','Code'])

enfermedades_totales.head()
```

```python
enfermedades_totales = pd.merge(enfermedades_totales, depresion_numero, on=['Entity','Year','Code'])

enfermedades_totales.head()
```

```python
world_population=pd.read_csv('Datos/salud mental/world_population_1990-2019.csv',index_col=0)
world_population.head()
```

```python
enfermedades_totales = pd.merge(enfermedades_totales, world_population, on=['Year'])

enfermedades_totales.head()
```

```python
enfermedades_totales.shape
```

```python
# Seleccionar las columnas numéricas a convertir
cols_to_convert = ['Mental_disorders_male(number)', 'Mental_disorders_female(number)', 'Total', 'Depressive_disorders(number)']

# divide cada columna numérica por la columna 'World Population' y multiplica por 100
enfermedades_totales[cols_to_convert] = enfermedades_totales[cols_to_convert].div(enfermedades_totales['World Population'], axis=0) * 100

# renombra las columnas para indicar que están en porcentaje
new_cols = [col.replace('(number)', '(%)') for col in cols_to_convert]
enfermedades_totales = enfermedades_totales.rename(columns=dict(zip(cols_to_convert, new_cols)))

enfermedades_totales
```
